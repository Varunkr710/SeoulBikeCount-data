**Dataset Overview**
The dataset used in this study contains several variables that are critical to understanding bike rental patterns. These variables include:

*   Date: The specific day on which the data was recorded.

*   Rented Bike Count: The total number of bikes rented on a given day or hour.

*   Hour: The hour of the day when the bike rentals were recorded.

*    Temperature (°C): The temperature at the time of bike rentals.

*   Humidity (%): The percentage of humidity in the air.

*   Wind Speed (m/s): The speed of the wind during the recording.

*   Visibility (10m): The distance over which objects can be clearly seen.

*   Dew Point Temperature (°C): The temperature at which air becomes saturated with moisture and dew forms.

*   Solar Radiation (MJ/m²): The amount of solar energy received per square meter.

*   Rainfall (mm): The amount of rainfall recorded.

*   Snowfall (cm): The amount of snowfall recorded.

*   Seasons: The season in which the data was recorded.

*   Holiday: Indicates whether the day was a holiday or not.

*   Functioning Day: Indicates whether the bike rental service was operational on that day.

**Key Findings and Analysis**
Our analysis revealed several key insights regarding the factors that influence bike rental counts:

1.   Time of Day: The data shows that the rented bike count is generally higher at night than during the day. This could be attributed to various factors such as reduced traffic, cooler temperatures, and the preference of users to rent bikes during off-peak hours.

2.   Weather Conditions: Weather plays a significant role in bike rentals. The analysis indicates that more bikes are rented on days without rainfall or snowfall. Adverse weather conditions like rain and snow discourage people from renting bikes, leading to a decrease in the rental count.

3.   Temperature: There is a direct correlation between temperature and bike rentals. As the temperature increases, the rented bike count also rises. This suggests that warmer weather is more conducive to biking, while colder temperatures may deter potential renters.

4.   Humidity: Humidity levels also impact bike rentals. The data shows that rentals are higher when humidity is moderate or low, but there is a significant drop in rentals when humidity levels are very high. High humidity can make outdoor activities uncomfortable, reducing the demand for bike rentals.

5.   Visibility and Wind Speed: Better visibility and higher wind speeds are associated with an increase in bike rentals. Clear visibility makes biking safer and more appealing, while wind speed might have a cooling effect, making the experience more comfortable.

6.   Seasonal Trends: Bike rentals peak during the summer months, specifically in June, July, and May. Conversely, the winter months of December, February, and January see a significant drop in bike rentals, likely due to colder temperatures and less favorable weather conditions.

7.   Day of the Week: The day of the week also influences bike rentals. Rentals are higher on Wednesdays and Fridays, while Sundays see the lowest rental counts. This pattern could be linked to commuting behaviors, with mid-week and end-of-week being popular times for bike usage.

8.   Average Rented Bike Count: The analysis shows that the average daily rented bike count can exceed 16,000 at a 5% or 10% confidence interval, but not at a 1% confidence interval. Similarly, the average hourly rented bike count does not exceed 700 at 1%, 5%, or 10% confidence intervals.

9.   Wind Speed and Humidity Relationship: There is no significant relationship between wind speed range and humidity range, as determined by the confidence intervals of 1%, 5%, and 10%.

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
def DateTomonth(x):        # convert date to month
   monthDictionary={'01':'January','02':'February','03':'March','04':'April','05':'May','06':'June','07':'July','08':'August','09':'September','10':'October','11':'November','12':'December'}
   return monthDictionary[x[3:5]]
def TemperatureRange(x):   # Dividing temperature to different range
  if x<3.5:
    return "low Temperature"
  elif x>=3.5 and x<22.5:
    return "Moderate Temperature"
  else:
    return "High Temperature"
def HumidityRange(x):    # Dividing humidity to differenet range
  if x<42:
    return "low"
  elif x>=42 and x<74:
    return "Moderate"
  else:
    return "High"
def WindSpeedRange(x):   # Dividing Wind Speed to differenet range
  if x<.9:
    return "low"
  elif x>=.9 and x<2.3:
    return "Moderate"
  else:
    return "High"
def VisibilityRange(x):    # Dividing Visibility to differenet range
  if x<940:
    return "low"
  elif x>=940 and x<1698:
    return "Moderate"
  else:
    return "High"
def dateToWeek(x):       # Convert date to week
  x=datetime.strptime(x,'%d/%m/%Y')
  return x.strftime("%A")

SeoulBikeData['Month']=SeoulBikeData['Date'].apply(DateTomonth)                                      # apply function to convert
SeoulBikeData['Day/Night']=SeoulBikeData['Hour'].apply(lambda x: "Day" if x<6 or x>19 else "Night")    # apply function to convert
SeoulBikeData['TemperatureRange']=SeoulBikeData['Temperature(°C)'].apply(TemperatureRange)               # apply function to convert
SeoulBikeData['HumidityRange']=SeoulBikeData['Humidity(%)'].apply(HumidityRange)                        # apply function to convert
SeoulBikeData['WindSpeedRange']=SeoulBikeData['Wind speed (m/s)'].apply(WindSpeedRange)                # apply function to convert
SeoulBikeData['VisibilityRange']=SeoulBikeData['Visibility (10m)'].apply(VisibilityRange)                 # apply function to convert
SeoulBikeData['Rainfall']=SeoulBikeData['Rainfall(mm)'].apply(lambda x : "no" if x==0 else "yes")           # apply function to convert
SeoulBikeData['Snowfall']=SeoulBikeData['Snowfall (cm)'].apply(lambda x : "no" if x==0 else "yes")        # apply function to convert
SeoulBikeData['WeekDay']=SeoulBikeData['Date'].apply(dateToWeek)                     # apply function to convert
SeoulBikeData
````

I have made several manipulations to the dataset including:

3.   Created two new columns, 'Rainfall' and 'Snowfall,' from the 'Rainfall (mm)' and 'Snowfall (cm)' columns, respectively. These columns indicate whether there was rainfall or snowfall during a particular hour.

4.   Introduced a 'Day/Night' column by converting hours 0 to 6 and 19 to 23 into 'Night' and the remaining hours into 'Day.'

5.   Created a 'TemperatureRange' column by categorizing the values in the 'Temperature (°C)' column into different ranges.

### Hypothetical Statement - 3

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

*   Null hypothesis:-There is a no relation between HumidityRange and WindspeedRange.
*   Alternate hypothesis:-There is a relation between HumidityRange and WindspeedRange.


#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
hypothesis3=SeoulBikeData[['HumidityRange','WindSpeedRange','Rented Bike Count']]
hypothesis3=hypothesis3.groupby(['HumidityRange','WindSpeedRange']).sum()
hypothesis3['Rented Bike Count']=(hypothesis3['Rented Bike Count']*100)/(hypothesis3['Rented Bike Count'].max()+hypothesis3['Rented Bike Count'].min())
def humid(x):
   value=hypothesis3[hypothesis3['HumidityRange']==x]
   return value['Rented Bike Count'].sum()
def wind(x):
   value=hypothesis3[hypothesis3['WindSpeedRange']==x]
   return value['Rented Bike Count'].sum()
hypothesis3=hypothesis3.reset_index()
hypothesis3['Humidsum']=hypothesis3['HumidityRange'].apply(humid)
hypothesis3['wind']=hypothesis3['WindSpeedRange'].apply(wind)
hypothesis3['E-values']=(hypothesis3['Humidsum']*hypothesis3['wind'])/(hypothesis3['Rented Bike Count'].sum())
hypothesis3['Square(Evalues)']=((hypothesis3['E-values']-hypothesis3['Rented Bike Count'])**2)/hypothesis3['E-values']
chi_square_value=hypothesis3['Square(Evalues)'].sum()
rows=len(hypothesis3['Humidsum'].unique())
columns=len(hypothesis3['wind'].unique())
df=(rows-1)*(columns-1)
print(stats.t.cdf(chi_square_value, df)*100)
````

##### Which statistical test have you done to obtain P-Value?

I have used chi-square test to calculate p-value.

I choose this specific test because it is the best method to calculate the relationship between two categorical variables.
